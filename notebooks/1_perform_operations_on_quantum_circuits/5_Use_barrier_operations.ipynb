qc.barier(0,1)

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1)
qc.h(0)
qc.barrier()
qc.h(0)
qc.draw('mpl')
```

```python
from qiskit import transpile, IBMQ
provider = IBMQ.load_account()
backend = provider.backend.ibmq_santiago

transpiled = transpile(qc, backend=backend)
transpiled.draw('mpl')
```

Remove barrier :

# Add Barrier from circuit library

```python
from qiskit.circuit.barrier import Barrier
qc = QuantumCircuit(1)
qc.h(0)
qc.append(Barrier(0),[0])
qc.h(0)
qc.draw('mpl')
```

Barrier over some random qubits

```python
qc = QuantumCircuit(5)
qc.barrier([0,3])
qc.draw('mpl')
```

```python
qc = QuantumCircuit(5)
qc.barrier(0,3)
qc.draw('mpl')
```
