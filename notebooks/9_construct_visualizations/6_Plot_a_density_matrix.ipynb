```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import DensityMatrix
from qiskit.visualization import plot_state_city
from qiskit.quantum_info import Statevector
%matplotlib inline
```

# Bell state 00

```python
bell_00 = QuantumCircuit(2)
bell_00.h(0)
bell_00.cx(0, 1)
bell_00.x(1)
print(bell_00)

state = DensityMatrix.from_instruction(bell_00)
print(state)
plot_state_city(state, color=['midnightblue', 'midnightblue'],
     title="New State City")
```

```python
sv = Statevector.from_label('00') #we used the .from_label(), we could also use .from_int(), for more check docs!
sv_ev = sv.evolve(bell_00) #we evolve this initial state through our circuit
sv_ev.draw('latex')
```
